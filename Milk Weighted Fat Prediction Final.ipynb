```python
import pandas as pd
data=pd.read_csv("Milk Fat SnF data 2 Months.csv")
```

```python
data.columns
```

```python
indep=data[['Fat','Weighted_Fat','Eraiyur_Qty','Naripalayam_Qty','Ellaigramam_Qty','Kadiyar_Qty','Kolathur_Qty']]
dep=data[['Eraiyur_WF','Naripalayam_WF','Ellaigramam_WF','Kadiyar_WF', 'Kolathur_WF']]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(indep,dep,test_size=1/3,random_state=0)
from sklearn.preprocessing import StandardScaler
sc_indep=StandardScaler()
x_train_sc=sc_indep.fit_transform(x_train)
x_test_sc=sc_indep.transform(x_test)

from sklearn.preprocessing import StandardScaler
sc_dep=StandardScaler()
y_train_sc=sc_dep.fit_transform(y_train)
y_test_sc=sc_dep.transform(y_test)
```

```python
from sklearn.multioutput import MultiOutputRegressor as MOR
from sklearn.linear_model import RidgeCV,Ridge
alpha=[0.1,1.0,10.0,100.0]
ridge_cv=MOR(RidgeCV(alphas=alpha,cv=5,scoring='neg_mean_squared_error'))
ridge_cv.fit(x_train_sc,y_train_sc)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error
y_pred=ridge_cv.predict(x_test_sc)
r=r2_score(y_test_sc,y_pred)
mae=mean_absolute_error(y_test_sc,y_pred)
print(r,mae)
```
